2) Importing all the libraries needed

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import *
from PIL import Image
import PIL.Image
from optparse import OptionParser
from pytesseract import pytesseract
from pytesseract import image_to_string
from cv2 import *
import numpy as np
```

3) Loading the Test Image and auto-adjustment of the pixel values

```
filename = "/content/drive/My Drive/business_card_folder/card/test_image.jpg"
im = cv2.imread(filename)
gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(1,1),1000)
flag, thresh = cv2.threshold(blur, 120, 255, cv2.THRESH_BINARY)
cv2_imshow(im)
```

```
numcards = 1
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=cv2.contourArea,reverse=True)[:numcards] 
```

```
for i in range(numcards):
  card = contours[i]
  peri = cv2.arcLength(card,True)
  approx = cv2.approxPolyDP(card,0.02*peri,True)
  rect = cv2.minAreaRect(contours[0])
  r = cv2.boxPoints(rect)
```

4) Reorientation of the Image (Image Segmentation and Crop)

```
approx = np.float32(approx)
h = np.array([ [0,0],[449,0],[449,449],[0,449] ],np.float32)
transform = cv2.getPerspectiveTransform(approx,h)
warp = cv2.warpPerspective(im,transform,(450,450))
warp = cv2.flip(warp,0)
cv2_imshow(warp)
```

5) Image Quality Enhancement

```
img = cv2.bilateralFilter(warp,100,90,90)
cv2_imshow(img)
```
